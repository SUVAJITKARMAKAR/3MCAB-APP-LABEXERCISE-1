```python
# Importing the modules in the workspace 

import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Importing the dataset and reading it the workspace

data_set = pd.read_csv('UScereal.csv')
```

```python
# Implementing bar-chart using 2 variables

plt.figure(figsize=(7,6))
plt.bar(data_set['mfr'], data_set['calories'], color='green')
plt.title('BAR-CHART USING 2 VARIABLES')
plt.xlabel('mfr')
plt.ylabel('calories')
plt.show()
```

```python
# Implementing bar-chart using 3 variables

sns.barplot(x='mfr', y='calories', hue='shelf', data=data_set)
plt.title('MANUFRATURER VS CALORIES WITH SHELF')
plt.show()
```

```python
# Implementing Scatterplots using 2 variables

plt.scatter(data_set['protein'], data_set['carbo'])
plt.title('PROTEIN VS CARBOHYDRATES')
plt.xlabel('protein')
plt.ylabel('carbohydrates')
plt.show()
```

```python
# Implementing Scatterplots using 3 varibles 

sns.scatterplot(x='protein', y='carbo', hue='shelf', data=data_set)
plt.title('PROTEIN VS CARBOHYDRATES WITH SHELF')
plt.show()
```

```python
# Implementing Histogram using 2 variables 

plt.hist(data_set['calories'], bins=20, color='crimson', edgecolor='black')
plt.title('CALORIES DISTRIBUTION')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
# Implementing Histogram using 3 variables

sns.histplot(x='calories', hue='mfr', data=data_set, bins=20, multiple='stack')
plt.title('CALORIES DISTRIBUTION BY MANUFRACTURER')
plt.show()
```

```python
# Implementing Pie-chart 

plt.pie(data_set['shelf'].value_counts(), labels=data_set['shelf'].value_counts().index, autopct='%1.1f%%', startangle=90)
plt.title('SHELF DISTRIBUTION')
plt.show()
```

```python
# Implementing the pair-chart

sns.pairplot(data_set[['calories', 'protein', 'carbo']])
plt.suptitle('CALORIES  PROTEIN  CARBOHYDRATES')
plt.show()
```

```python
# Implementing bubble-plot 

plt.scatter(data_set['protein'], data_set['carbo'], s=data_set['calories']*2, alpha=0.5, color='teal', edgecolors='black')
plt.title('PROTEIN VS CARBOHYDREATES WITH CALORIES')
plt.xlabel('Protein')
plt.ylabel('Carbohydrates')
plt.show()
```

```python
# Implementing box-plot 

sns.boxplot(x='mfr', y='calories', data=data_set, color='yellow', linecolor='black')
plt.title('MANUFACTURER VS CALORIES')
plt.show()
```

```python
# Implementing violen-plot

sns.violinplot(x='mfr', y='calories', data=data_set, color='orange')
plt.title('MANUFACTURER VS CALORIES')
plt.show()
```

```python
# Implementing area-plot 

data_set[['protein', 'carbo']].plot.area(stacked=False)
plt.title('PROTEIN VS CARBOHYDRATES')
plt.show()
```

```python
# Implementing heat-map

numeric_data_set = data_set.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numeric_data_set.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('CORRELATION MAP')
plt.show()
```
